```python
import numpy as np
import pandas as pd

scores = np.random.randint(40,high=101,size=(50,5))
scores_df=pd.DataFrame(scores,
    columns=['國文','英文','數學','地理','歷史'],
    index=range(1,51))
scores_df.index.name="學號"
scores_df.columns.name="科目"

#display(sum_series)               #sum總分 #mean平均 #rank排名
sum_series=scores_df.sum(axis=1)
mean_series=scores_df.mean(axis=1)
rank_series=sum_series.rank(method='min',ascending=False) #rank依照總分1~50(但表格整體排列沒有變)
scores_df['總分']=sum_series
scores_df['平均']=mean_series
scores_df['名次']=rank_series
scores_df.sort_values(by="名次")   #sort_value把表格依照名次排列
```

```python
def fun1(series:pd.Series)-> int:
    return series.max() #傳出每個series的最高分：國文最高分、英文最高分、數學最高分、地理最高分、歷史最高分

subject_df=scores_df[['國文','英文','數學','地理','歷史']]

subject_df.apply(fun1)
```

```python
def fun1(series:pd.Series)-> int:
    s1 = pd.Series(series.median(),index=['中間值'])  #傳出中間值
    return s1 #傳出s1

subject_df=scores_df[['國文','英文','數學','地理','歷史']]

subject_df.apply(fun1)
```

```python
def fun1(series:pd.Series)-> int:
    min=series.min()         #先定義min
    median=series.median()   #先定義median
    max=series.max()         #先定義max
    s1=pd.Series([min,median,max],index=['最低分','中間值','最高分'])  #傳出最低分中間值
    return s1 #傳出s1

subject_df=scores_df[['國文','英文','數學','地理','歷史']]

frame1=subject_df.apply(fun1)
frame1
```

```python
def fun2(series:pd.Series) ->pd.Series:
    lessthan60=series[series<60].count()                    #定義
    chinese ='及格' if (series['國文'] >=60) else "不及格"   #定義
    english ='及格' if (series['英文'] >=60) else "不及格"   #定義
    
    return pd.Series([lessthan60,chinese,english],index=['不及格科目','國文','英文'])

display(subject_df.T)
subject_df.T.apply(fun2).T   # .T欄列對調
```
