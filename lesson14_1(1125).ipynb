```python
import numpy as np
import pandas as pd
scores=np.random.randint(50,100,size=(50,5)) 
scores_df=pd.DataFrame(scores,
            columns=['國文','英文','數學','地理','歷史'],
            index=range(1,51))
scores_df
sum_values=scores_df.sum(axis=1)
mean_values=scores_df.mean(axis=1)
scores_df['總分']=sum_values
scores_df['平均']=mean_values
```

```python
scores_df.sort_values(by='總分',ascending=False).iloc[:5] #ascending由小變大，#ascending=False由大變小
```

```python
#使用subscript[],過濾資料
scores_df[scores_df['平均']<80]   #series陣列運算可用加減乘除，比較運算子<產生布林值 true、false
```
